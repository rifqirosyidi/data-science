```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### **1. Sample Plot with 4 data**

```python
plt.plot([1, 3, 2, 4])
plt.show()
```

#### **Test Plot**

```python
plt.plot([1, 2, 3, 4], [1, 9, 10, 18])
plt.title("TEST PLOT", pad=15, fontsize=14, color='red')
plt.xlabel("First Arr")
plt.ylabel("Second Arr")
plt.show()
```

#### **N Pangkat**

```python
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25])
plt.title("N PANGKAT", pad=15, fontsize=14, color='red')
plt.xlabel("n")
plt.ylabel("n^2")
plt.show()
```

### **2. Change The Size**

##### **X wide = 2; Y wide = 5**

##### **with X axis = 0-5; Y axis = 0-20**

```python
plt.figure(figsize=(2,5))
plt.plot([1,2,3,4], [1,4,9,16], 'bx')  # red=r; black=x; circle=o; cross=x; 
plt.axis([0, 6, 0, 20])  # Axis [xstart, xend, ystart, yend]
plt.annotate("cool right", (3,7))
plt.show()
```

### **3. Bar Chart**

```python
plt.clf() # Clear Figure
x = np.arange(6)
y = np.random.randint(1, 10, 6)
plt.xticks(x, ('Data A', 'Data B', 'Data C', 'Data D', 'Data E', 'Data F'))
plt.bar(x, y, color=['g', 'y'])  # color : y, k, b, r, lime, tan *** OR *** color [t, r, b] this will give different color
plt.show()
```

### **4. Random Dots**

```python
data = {
    'Blue O': np.random.rand(10),
    'Red X': np.random.rand(10)
}

data_frame = pd.DataFrame(data)
print(data_frame)

data_frame.plot()

data_frame.plot(style=['bo', 'rx'])

plt.show()
```

### **4. Time Series**

```python
time_stamp = pd.Series(np.random.randn(180), index=pd.date_range('1/1/2019', periods=180))
data_frame = pd.DataFrame(np.random.randn(180,3), index=time_stamp.index, columns=['A','B','C'])  # or columns=list(ABC) > same

print(data_frame)
print(data_frame.cumsum())

data_frame.cumsum().plot()
plt.show()
```

### **5. Scatter**

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
sizes = (30 * np.random.rand(N)) ** 2
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
plt.show()
```

### **6. Load CSV to Multiple Data Chart**

```python
data_frame = pd.read_csv('raw_data.txt')

plt.bar(data_frame.month, data_frame.record_high, color='k')
plt.bar(data_frame.month, data_frame.record_low, color='grey')
plt.plot(data_frame.avg_high)
plt.plot(data_frame.avg_low)
plt.legend(['avg_high', 'avg_low', 'record_high', 'record_low'])
plt.show()
```

### **7. SubPlot**

```python
fig = plt.figure()
fig.suptitle("My Sub Plot")
fig.add_subplot(221)   #top left
plt.plot([1,2,3])
fig.add_subplot(222)   #top right
plt.plot([1,3,2,4])
fig.add_subplot(223, facecolor='y')   #bottom left
fig.add_subplot(224)   #bottom right 
plt.show()
```

### **8. Subplot II**

```python
fig, plots = plt.subplots(2, sharex=True)
fig.suptitle("Sharing X")
x = range(0, 100)
y = [n/2 for n in x]
print(x)
plots[0].plot(x,y, color="r")
plots[1].plot(x,y)
plt.show()
```

### **9. Saving the Plot**

```python
plt.figure(figsize=(4,3), dpi=100)
plt.plot([145, 178, 147, 164, 177, 143, 132, 198, 177])
plt.axis([0,8, 0, 250])
plt.title("Flight Data")
plt.xlabel("Speed")
plt.savefig("flight.png")
plt.show()
```
