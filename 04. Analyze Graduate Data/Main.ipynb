```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### **1. Initial Read Data**

```python
df = pd.read_csv("data_to_predict.csv")
print(df.shape)
print(df.info())
print(df)
```

### **2. Cleaning The Data**

```python
df = df.drop(['Serial No.'], axis=1)
new_names = {'GRE Score':'gre', 'TOEFL Score':'toefl', 'University Rating':'rating', 'SOP':'sop', 'LOR':'lor', 'CGPA':'gpa', 'Research':'research', 'Chance of Admit ':'chance'}
df.rename(columns=new_names, inplace=True)
df.head()
```

### **3. Subplot Fig**

```python
plt.figure(1, figsize=(15, 4))
for i in range(1, 6):
    plt.subplot(1, 5, i)
    plt.boxplot([df[df.columns[i]]])
    plt.title(df.columns[i])
plt.show()
df.describe()
```

### **4. Detailed Data Distribution**

```python
df.hist(figsize=(20,10), bins=16, edgecolor='k', grid=False)
plt.show()
```

### **5. Correlation**

```python
correlation = df.corr()['chance']
plt.bar(df.columns, correlation)
plt.show()
```

### **6. Relation**

```python
df.plot(kind='scatter', x='gre', y='toefl', color='green')
plt.xlabel('GRE')
plt.ylabel('TOEFL')
plt.show()
```

GRE High == Toefl High, GRE low == Toefl Low

```python
df.plot(kind='scatter', x='gre', y='gpa', color='blue')
plt.xlabel('GRE')
plt.ylabel('GPA')
plt.show()
```

### **7. Correlation between Chance and other fields**

```python
for i in range(1,7):
    df.plot(kind='scatter', x='chance', y=df.columns[i], color='blue')
    plt.xlabel('GRE')
    plt.ylabel(df.columns[i])
    plt.show()
```

__what we want to see is less spead of a point of a cluster, look on the data above the gpa and toefl is more likely the field which increase the percentage chance__
